```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('C:/Users/ASUS/Machine_Learning/100_Days_Of_Deep_Learning/Datasets/placement(1).csv')
```

```python
df.head()
```

```python
df.shape
```

```python
sns.scatterplot(df,x = df.cgpa,y = df.resume_score,hue = df.placed)
plt.show()
```

```python
X = df.iloc[:,:2]
y = df.iloc[:,-1]
```

```python
from sklearn.linear_model import Perceptron

p = Perceptron()
```

```python
p.fit(X,y)
```

```python
p.coef_
```

```python
p.intercept_
```

```python
from mlxtend.plotting import plot_decision_regions

plot_decision_regions(X.values,y.values,clf=p,legend=2)


plt.xlabel('cgpa')
plt.ylabel('resume_score')
plt.title('Perceptron On Placement')
plt.show()
```
