```python
import pandas as pd
```

```python
df = pd.read_csv("/content/Customer Call List.xlsx - Call List.csv")
df
```

```python
# Removing Duplicates
df = df.drop_duplicates()
df
```

```python
# df.drop("Not_Useful_Column", axis = columns)
df = df.drop(columns = "Not_Useful_Column")
df
```

```python
df["Last_Name"] = df["Last_Name"].str.lstrip("/")
df["Last_Name"] = df["Last_Name"].str.lstrip("...")
df["Last_Name"] = df["Last_Name"].str.rstrip("_")
# df["Last_Name"] = df["Last_Name"].str.strip("123_./")
df
```

```python
df["Phone_Number"] = df["Phone_Number"].str.replace("[^a-zA-Z0-9]","")
df
```

```python
df["Phone_Number"] = df["Phone_Number"].apply(lambda x : str(x))
'''
Using Foor Loop
for i in range(len(df["Phone_Number"])):
    df["Phone_Number"][i] = str(df["Phone_Number"][i])
print(type(df["Phone_Number"][0]))
'''
```

```python
df["Phone_Number"] = df["Phone_Number"].apply(lambda x : x[0:3] + "-" + x[3:6] + "-" + x[6:10])
'''
#Using Foor Loop
for i in range(len(df["Phone_Number"])):
    df["Phone_Number"][i] = df["Phone_Number"][i][0:3] + "-" + df["Phone_Number"][i][3:6] + "-" + df["Phone_Number"][i][6:10]
print(type(df["Phone_Number"][0]))
'''
df
```

```python
df["Phone_Number"] = df["Phone_Number"].str.replace("nan--", "")
df["Phone_Number"] = df["Phone_Number"].str.replace("Na--", "")
df
```

```python
df[["Street_Address","State","Zip_Code"]] = df["Address"].str.split("," , 2 , expand = True)
df
```

```python
df = df.drop("Address", axis='columns')
```

```python
df["Paying Customer"] = df["Paying Customer"].str.replace("Yes", "Y")
df["Paying Customer"] = df["Paying Customer"].str.replace("No", "N")
df["Paying Customer"]
```

```python
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace("Yes", "Y")
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace("No", "N")
df["Do_Not_Contact"]
```

```python
df = df.fillna('')
df = df.replace('N/a', '')
df
```

```python
for i in df.index:
    if df.loc[i,"Do_Not_Contact"] == "Y":
        df.drop(i , inplace = True)
df
```

```python
for i in df.index:
    if df.loc[i,"Phone_Number"] == "":
        df.drop(i , inplace = True)
df

# df.dropna(subset = "Phone_Number", inplace = True)  another way to drop null values
```

```python
df = df.reset_index(drop = True)
df
```
