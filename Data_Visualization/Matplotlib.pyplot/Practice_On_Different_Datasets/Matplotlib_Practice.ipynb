# Subplot

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
y4 = np.exp(x)


fig, axs = plt.subplots(2, 2, figsize=(15,10))

axs[0, 0].scatter(x, y1, color='r')
axs[0, 0].set_title('Sin(x)')

axs[0, 1].scatter(x, y2, color='g')
axs[0, 1].set_title('Cos(x)')

axs[1, 0].scatter(x, y3, color='b')
axs[1, 0].set_title('Tan(x)')

axs[1, 1].scatter(x, y4, color='m')
axs[1, 1].set_title('Exp(x)')

plt.tight_layout()
plt.show()
```

```python
x = np.arange(-5, 6, 1)
y = np.arange(-5, 6, 1)

X, Y = np.meshgrid(x, y)

Z_surface = np.sin(np.sqrt(X**2 + Y**2))

fig = plt.figure(figsize=(12, 6))

ax1 = fig.add_subplot(121, projection='3d')
ax1.plot_surface(X, Y, Z_surface, cmap='viridis')
ax1.set_title('3D Surface Plot')
ax1.set_xlabel('X-axis')
ax1.set_ylabel('Y-axis')
ax1.set_zlabel('Z-axis')

ax2 = fig.add_subplot(122)
contour = ax2.contourf(X, Y, Z_surface, cmap='viridis')
ax2.set_title('Contour Plot')
ax2.set_xlabel('X-axis')
ax2.set_ylabel('Y-axis')
plt.colorbar(contour, ax=ax2)

plt.tight_layout()
plt.show()
```
